```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import seaborn as sns
import numpy as np

# 读取本地数据文件
df = pd.read_excel(r'C:\Users\Jennie\Desktop\jiaoshi.xlsx', header=0)

# 选择需要进行聚类的列（'AGE'）
selected_column = 'AGE'
df = df.dropna(subset=[selected_column])

# 标准化数据
scaler = StandardScaler()
df[selected_column] = scaler.fit_transform(df[[selected_column]])

# 使用KMeans进行聚类
n_clusters = 2  # 请根据实际情况选择聚类的数量
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(df[[selected_column]])

# 打印聚类结果
print(df[['AGE', 'cluster']])

# 可视化展示
# 1. 散点图
plt.scatter(df.index, df['AGE'], c=df['cluster'], cmap='rainbow')
plt.xlabel('Index')
plt.ylabel('AGE')
plt.title('Scatter Plot of Clusters')
plt.show()

# 2. 箱线图
sns.boxplot(x='cluster', y='AGE', data=df)
plt.xlabel('Cluster')
plt.ylabel('AGE')
plt.title('Boxplot of AGE by Cluster')
plt.show()

# 3. 直方图
for cluster_label in df['cluster'].unique():
    plt.hist(df[df['cluster'] == cluster_label]['AGE'], label=f'Cluster {cluster_label}', alpha=0.5)

plt.xlabel('AGE')
plt.ylabel('Frequency')
plt.title('Histogram of AGE by Cluster')
plt.legend()
plt.show()
```
