```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns

# 读取本地数据文件
df = pd.read_excel(r'C:\Users\Jennie\Desktop\jiaoshi.xlsx', header=0)

# 选择需要进行聚类的列（'GENDER'）
selected_columns = ['GENDER']
df = df.dropna(subset=selected_columns)

# 处理缺失值
df = df.dropna(subset=selected_columns)

# 选择需要进行聚类的列（'GENDER'）
selected_columns = ['GENDER']

# 使用 KMeans 进行聚类
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(df[selected_columns])

# 可视化展示
# 1. 散点图
plt.scatter(df.index, df['GENDER'], c=df['cluster'], cmap='rainbow')
plt.xlabel('Index')
plt.ylabel('GENDER')
plt.title('Scatter Plot of Clusters (by Gender)')
plt.show()

# 2. 箱线图
sns.boxplot(x='cluster', y='GENDER', data=df)
plt.xlabel('Cluster')
plt.ylabel('GENDER')
plt.title('Boxplot of GENDER by Cluster')
plt.show()

#3.直方图
for cluster_label in df['cluster'].unique():
    plt.hist(df[df['cluster'] == cluster_label]['GENDER'], label=f'Cluster {cluster_label}', alpha=0.5)

plt.xlabel('GENDER')
plt.ylabel('Frequency')
plt.title('Histogram of GENDER by Cluster')
plt.legend()
plt.show()
```
